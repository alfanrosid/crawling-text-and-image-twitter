```python
import tweepy # libarary untuk Twitter
import csv    # library untuk membuat file CSV
import urllib.request
import requests as r
```

```python
# Inisialisasi API Key yang terdaftar
api_key = ''
api_key_secret = ''
access_token = ''
access_token_secret = ''
```

```python
# Autentikasi API Twitter menggunakan Tweepy
auth = tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)
search_key="kadrun"
        
def save_to_files(url_images, name_files):
    open_url = r.get(url_images)
    open("datameme//"+name_files, 'wb').write(open_url.content)
```

```python
#membuat file csv
csvFile = open(search_key+".csv","a+", newline="", encoding="utf-8")
csvWriter = csv.writer(csvFile)

i=1
for tweet in tweepy.Cursor(api.search,q=search_key+" -filter:retweets",count=1000,lang="id",since="2021-10-07",until="2021-10-08").items(): # cari tweet dengan search_key yang berbahasa indonesia dengan 10 tweet per halaman dari tanggal xxxxx 
    print(tweet.created_at,tweet.id,tweet.user.name,tweet.text)
    
    try:
        do = api.get_status(tweet.id)._json
        url_images = do['entities']['media'][0]['media_url_https']
        fullname = "meme"+str(i)+".jpg"
        save_to_files(url_images, fullname)
        #mengisi csv
        tweets=[tweet.created_at,tweet.id,tweet.user.name,tweet.text.encode("utf-8"),fullname]
        csvWriter.writerow(tweets)
        i=i+1
    except:
        continue

```
